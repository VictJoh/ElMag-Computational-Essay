## Introduksjon

Vi skal utforske hvordan en mikroblgeovn fungerer i elektromagnetisk kontekst. Dette er spesifikt for  kunne se hvor raskt og hvordan en mikroblgeovn kunne drept en fisk avhengig av strrelse og utforming p fisken
For  gjre eksperimentet mer humant skal vi anta at fisken er sedert og dermed ikke ved bevissthet. Dette gjr det i tillegg lettere ettersom legemet vil ligge i ro under prosessen.

## Teori

Gjennomsnittet av tall funnet p nettet med litt buffer her og der gir 22 minutter (1320 sek) overlevelsestid for en laks p land. Ettersom generelle tall er vanskelig  finne, sier vi at den gjennomsnittlige laksen overlever i den gjennomsnittlige tiden, og dette avtar/ker linert avhengig av strrelsen p fisken, da bestemt  av vekt.

I de kommende likningene vil disse uttrykkene bli representert av tilhrende bokstaver:

$$
\text{Levetid} = t
$$ 
$$
\text{Strrelse\_p\_fisk} = m
$$
$$
\text{Konstant} = k, \;\;\;\;\;\text{$k$ er forholdet mellom strrelse og levetid}
$$

Med dette kan vi beskrive overlevelsestiden ved hjelp av likningen:

$$
t = k \cdot m
$$

Vi beregner da overlevelseskonstanten $k$ ved  snu p formelen, og fr:

$$
k = \frac{\text{Levetid}}{\text{strrelse\_p\_fisk}} = \frac{t}{m} = \frac{1320 \, \text{sek}}{4.5 \, \text{kg}} = \frac{880}{3} \, \frac{\text{sek}}{\text{kg}}
$$

Vi antar dermed at den gjennomsnittlige laksen, omtrent 4.5 kg, overlever i 22 minutter, og at nr strrelsen avtar, avtar ogs overlevelsestiden.

$$
\text{Levetid} = t = k \cdot m 
$$

Dermed vil en fisk p 30 kg vre antatt  overleve i 8800 sek eller 146.66 minutter.

Dette er ved ganske ideelle leveforhold i forhold til en mikroblgeovn. Derfor antar vi at en fisk i mikroblgeovn vil d $\sqrt{t}$ fortere enn forholdene brukt tidligere. Alts for en fisk p 4.5 kg:

$$
\text{ovn\_levetid} = \sqrt{t} = \sqrt{1320} \approx 36.3 \, \text{sek}
$$

Til sammenligning, fisk p 30 kg:

$$
\text{ovn\_levetid} = \sqrt{8800} \approx 93.8 \, \text{sek}
$$

Det var nok om fisken. N til hvordan en mikroblgeovn fungerer.

Mikroblgeovnen bruker oftest en frekvens p 2.45 GHz, og ovnen vi ser for oss har varmer fisken med en effekt p mellom 600 W til 1000 W, s dette er omrdet vi nsker  underske.

#### Modellering av fisken

$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2}  = 1
$$

#### Utregning for nr fisken dr av mikroblgeovnen    

Vi antar at blgene varmer utenfra og inn, alts at den ikke varmer p innsiden fr utsiden er varm. Vi antar ogs at en fisk kan sees som dd dersom $\frac{1}{10}$ av fiskenes bredde er varmet. Etter litt sking her og der lander den gjennomsnittlige bredden p en norsk laks p omtrent 14 cm, som vi skal anta er nyaktig heretter. Det vil si at nr mikroblgeovnen har varmet fisken $1.4 \, \text{cm}$ inn i kjttet, vil vi anse fisken som dd. Fisken starter p en temperatur p 10 grader Celsius, og vi anses ddelig temperatur som 40 grader Celsius. Dvs. fisken m varmes $30$ grader for  vre ddelig.

En trenger litt under $4200 \, \text{J}$ ($4186 \, \text{J}$) for  varme 1 kg vann med 1 grad. Siden fisken lever i vann, vil massetettheten ikke vre helt ulik vann, men den er litt tettere. For  gjre livet vrt enklere, antar vi dermed at vi trenger $4200 \, \text{J}$ for  varme 1 kg fisk med en grad, eller $126{,}000 \, \text{J}$ for  varme 1 kg fisk til 40 grader, alts ddelig temperatur.

N gjenstr det kun  regne ut hvor mange kg $\frac{1}{10}$ av fiskens bredde tilsvarer.

##### Hvordan fungerer mikroblgeovn
En mikroblgeovn opererer ved hjelp av en magnetron som omdanner elektrisk energi til blger. Det gjr den ved  f elektroner til  bevege seg i sirkulre baner som lager et indusert oscillerende magnetisk felt. Dette feltet vil generere mikroblger som vi kan bruke til varme. 

$$
\lambda = \frac{c}{f}
$$


#### Elektrisk felt

Dielektriske egenskaper:
I lufta fra mikroblgeovnen til den nr fisken vil det vre noe tap av styrke. Vi lar mikroblgeovnen emitere et felt med styrke $E_0$ som gr gjennom lufta som vi anser som vakuum og ser derfor for oss at vi har lite tap av elektrisk felt. Da har vi at $E(z) = E_0$ fr blgene har truffet fisken.

I fisken:
Vi har en dielektrisk konstant ($\epsilon_r$) inne i fisken som mler evnen til  ta opp elektrisk energi. \
Vi har ogs en tapfaktor kjent som $\tan{\delta}$ som representerer hvor effektivt materialet omdanner elektrisk energi til varme.

Siden vi antar at fisken bestr av mye vann s bruker vi dielektrisk konstant og tapfaktor til vann. For  forenkle bruker vi empiriske verdier ved romtemperatur ($20^\circ - 25^\circ$). Det er gitt ved $\epsilon_r = 78$ og $\tan\delta = \frac{Lf}{\epsilon_r}$, hvor $Lf$ er tapsfaktor (Loss factor) ofte gitt som $\epsilon''$ og er den imaginre delen av av den dielektriske konstanten. Dette har vi at er omtrent $Lf = 12$ fra empirisk data ved 2.45 GHz. Det gir $\tan\delta = \frac{12}{78} = 0.153846$ https://water.lsbu.ac.uk/water/microwave_water.html

Elektrisk felt:
Styrken p feltet vil da kun avhenge av hvor langt det har brutt inn i fisken som vi gir ved $r$. Dersm $E_0$ er styrken ved $r=0$ p overflaten fr vi flgende:

$$
E(r) = E_0e^{-\alpha r}
$$
Hvor $\alpha$ er attenueringskonstanten gitt ved:

$$\alpha = \frac{\sqrt{2\epsilon_r}\pi f \tan{\delta}}{c} = \sqrt{2}\pi \frac{l_f}{\lambda\sqrt{\epsilon_r}}$$


Grunnen til at vi bruker $r$ n og ikke $z$ er at feletet vil n endre retning ettersom det endrer permitivitet og fisken er formet som en ellipsoide som gjr at det vil danne en innfalsvinkel. Dermed vil ikke feltet lenger virke kun i $z$-retning.



#### Oppvarming

Vi nsker  finne $P$ som er energien som blir omdannet til varme. Den er gitt ved:
$$ P = 2 \pi f \epsilon_0 l_f E^2(r) = 2\pi \frac{c}{\lambda}\epsilon_0 l_f E_0^{2}e^{-\sqrt{2}\pi \frac{l_f}{\lambda\sqrt{\epsilon_r}}r}$$
 Som vi ser s er vi blitt tildelt et vakkert stykke som er behagelig for ynene, men om man ser litt nrmere etter ser vi at det kun er noen f ting vi  mangler. 
 



## Antagelser under modellering

Nr mikroblgeovnen er p vil platen fisken ligger p rotere. For  forenkle systemet mest mulig vil vi anta at mikroblgeovnen kun avslutter etter hele rotasjoner av fisken, varmen da vil f en bedre form for symmetri over fisken. Vi antar ogs at mikroblgeovnen er en prisme som emiterer konstante mikroblger. 

Mikroblgeovnen lar vi ha en kvadratisk bunn med sider $40$ cm og hyde $10$ cm.

```python
import numpy as np
from scipy import constants
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets
from IPython.display import display
```

```python
mu_0 = constants.mu_0
epsilon_0 = constants.epsilon_0
c = 3e8   
```

```python
# mikroblgeovn
f = 2.45e9 # frekvens p mikroblger 2.45 GHz
lamb = c / f # blgelengde p mikroblger
P = 800 # effekt
A = 0.4 * 0.4 # areal av toppen
T = 1 / f
omega = 2 * np.pi * f  # vinkelhastighet


E0 = np.sqrt((2*P) / (epsilon_0 * c* A))
E0
```

```python
#fisk
epsilon_r = 78
l_f = 12
tan_delta = l_f / epsilon_r
alpha = omega * np.sqrt(mu_0 * epsilon_0 * epsilon_r / 2) * tan_delta
delta = 1 / alpha
```

```python
r = np.linspace(0, 0.10, 500)  
E = 1 * np.exp(-alpha * r)

plt.figure(figsize=(10,6))
plt.plot(r*100, E)
plt.title('Relativ Elektrisk Feltstyrke inne i fisken')
plt.xlabel('Dybde (cm)')
plt.ylabel('Elektrisk Feltstyrke (V/m)')
plt.grid(True)
plt.show()
```

```python
# definere ellipsoide
a_ellipsoid = 0.35 / 2    
b_ellipsoid = 0.07 / 2
c_ellipsoid = 0.05 / 2
```

```python
num_points = 200

# polar koordinater - meshgrid
u = np.linspace(0, 2 * np.pi, num_points)
v = np.linspace(0, np.pi, num_points)
U, V = np.meshgrid(u, v)

X = a_ellipsoid * np.cos(U) * np.sin(V)
Y = b_ellipsoid * np.sin(U) * np.sin(V)
Z = c_ellipsoid * np.cos(V)

# plot ellipsoide
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='pink', alpha=0.6, edgecolor='pink')

# strrelse p mikroblgeovn
ax.set_xlim([-0.2, 0.2]) 
ax.set_ylim([-0.2, 0.2])
ax.set_zlim([0, 0.1])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('Ellipsoid Plot')
plt.show()
```

```python
    # nye konstanter

mu_r = 1.0 # Fisk er ikke magnetisk

# Refraktiv indeks til bruk for Snells lov
n_air = 1.0
n_fish = np.sqrt(epsilon_r * mu_r)
k_air = omega / c  
k_fish = n_fish * omega / c  
```

```python

t0 = 0
t1 = 0.5 * T

num_points = 200
x = np.linspace(-0.20, 0.20, num_points)
z = np.linspace(0, 0.10, num_points)
X, Z = np.meshgrid(x, z)
Y = 0  
R_ellipsoid = (X / a_ellipsoid)**2 + (Y / b_ellipsoid)**2 + ((Z - c_ellipsoid) / c_ellipsoid)**2
inside_fish = R_ellipsoid <= 1

# Calculate the distance from the surface along the Z-axis
distance_inside = np.zeros_like(X)
Z_surface = c_ellipsoid + c_ellipsoid * np.sqrt(1 - (X[inside_fish] / a_ellipsoid)**2)
distance_inside[inside_fish] = Z_surface - Z[inside_fish]
distance_inside = np.maximum(distance_inside, 0)

# Initialize electric field components
E_z = np.zeros_like(Z)
E_x = np.zeros_like(X)

# Calculate the normal vector components at the surface
n_x = np.zeros_like(X)
n_z = np.zeros_like(Z)
n_x[inside_fish] = X[inside_fish] / a_ellipsoid**2
n_z[inside_fish] = (Z_surface - c_ellipsoid) / c_ellipsoid**2

# Calculate the angle of incidence and transmission using Snell's Law
# Calculate the angle of incidence using the normalized normal vectors
theta_i = np.arctan2(n_x[inside_fish], n_z[inside_fish])

# Calculate the sine of the transmission angle using Snell's Law
sin_theta_t = (n_air / n_fish) * np.sin(theta_i)

# Initialize theta_t with NaNs
theta_t = np.full_like(theta_i, np.nan)

# Identify where total internal reflection does not occur
valid = np.abs(sin_theta_t) <= 1

# Calculate theta_t where valid
theta_t[valid] = np.arcsin(sin_theta_t[valid])

# Loop over times t0 and t1
t_values = [t0, t1]

for t in t_values:
    # Outside the fish (in air)
    E_z_outside = E0 * np.cos(k_air * Z - omega * t)
    E_x_outside = np.zeros_like(X)

    # Inside the fish
    E_amplitude_inside = E0 * np.exp(-alpha * distance_inside[inside_fish])

    # Initialize E_z_inside and E_x_inside
    E_z_inside = np.zeros_like(distance_inside[inside_fish])
    E_x_inside = np.zeros_like(distance_inside[inside_fish])

    # Only compute fields where theta_t is valid
    valid = np.abs(sin_theta_t) <= 1

    # Compute phase inside the fish
    E_phase_inside = k_fish * (X[inside_fish] * np.sin(theta_t) + Z[inside_fish] * np.cos(theta_t)) - omega * t

    # Assign electric field components where valid
    E_z_inside[valid] = E_amplitude_inside[valid] * np.cos(theta_t[valid]) * np.cos(E_phase_inside[valid])
    E_x_inside[valid] = E_amplitude_inside[valid] * np.sin(theta_t[valid]) * np.cos(E_phase_inside[valid])


    E_z[~inside_fish] = E_z_outside[~inside_fish]
    E_x[~inside_fish] = E_x_outside[~inside_fish]
    E_z[inside_fish] = E_z_inside
    E_x[inside_fish] = E_x_inside

    # Compute the magnitude and normalize for quiver plot
    magnitude = np.sqrt(E_x**2 + E_z**2)
    Ex_norm = E_x / magnitude
    Ez_norm = -E_z / magnitude  # Adjust direction for plotting

    # Plotting the electric field strength at time t
    plt.figure(figsize=(10, 8))
    contour = plt.contourf(X, Z, E_z, levels=50, cmap="viridis")
    plt.colorbar(contour, label="Electric Field Strength (V/m)")
    plt.xlabel("X (m)")
    plt.ylabel("Z (m)")
    plt.title(f"Electric Field Distribution and Direction at t = {t:.2e} s")

    # Plot the ellipsoid boundary
    theta_boundary = np.linspace(0, 2 * np.pi, 300)
    x_ellipsoid = a_ellipsoid * np.cos(theta_boundary)
    z_ellipsoid = c_ellipsoid + c_ellipsoid * np.sin(theta_boundary)
    plt.plot(x_ellipsoid, z_ellipsoid, color='white', linewidth=2, label="Fish")

    # Plot electric field direction using quiver
    n_quiver = 20
    plt.quiver(X[::n_quiver, ::n_quiver], Z[::n_quiver, ::n_quiver],
               Ex_norm[::n_quiver, ::n_quiver], Ez_norm[::n_quiver, ::n_quiver],
               color='white', pivot='mid', scale=25, headwidth=3, headlength=4)

    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
```

```python
def P_total(Ex, Ez):
    epsilon_r_imag = epsilon_r * tan_delta
    E_total_squared = Ex**2 + Ez**2
    P_r = 0.5 * omega * epsilon_0 * epsilon_r_imag * E_total_squared
    return P_r
```

```python
import numpy as np
import matplotlib.pyplot as plt

# LAGE MASSE FELT :D


# Generate 100 time values from 0 to T
t_values = np.linspace(0, T, 100)
num_times = len(t_values)

# Spatial grid parameters
num_points = 200
x = np.linspace(-0.20, 0.20, num_points)
z = np.linspace(0, 0.10, num_points)
X, Z = np.meshgrid(x, z)
Y = 0

# Define the ellipsoid representing the fish
R_ellipsoid = (X / a_ellipsoid)**2 + (Y / b_ellipsoid)**2 + ((Z - c_ellipsoid) / c_ellipsoid)**2
inside_fish = R_ellipsoid <= 1

# Calculate the distance from the surface along the Z-axis
distance_inside = np.zeros_like(X)
Z_surface = c_ellipsoid + c_ellipsoid * np.sqrt(1 - (X[inside_fish] / a_ellipsoid)**2)
distance_inside[inside_fish] = Z_surface - Z[inside_fish]
distance_inside = np.maximum(distance_inside, 0)

# Calculate the normal vector components at the surface
n_x = np.zeros_like(X)
n_z = np.zeros_like(Z)
n_x[inside_fish] = X[inside_fish] / a_ellipsoid**2
n_z[inside_fish] = (Z_surface - c_ellipsoid) / c_ellipsoid**2

# Calculate the angle of incidence and transmission using Snell's Law
theta_i = np.arctan2(n_x[inside_fish], n_z[inside_fish])
sin_theta_t = (n_air / n_fish) * np.sin(theta_i)

# Initialize theta_t with NaNs
theta_t = np.full_like(theta_i, np.nan)

# Identify where total internal reflection does not occur
valid = np.abs(sin_theta_t) <= 1

# Calculate theta_t where valid
theta_t[valid] = np.arcsin(sin_theta_t[valid])

# Pre-allocate arrays to store E_x and E_z at each time step
E_x_all = np.zeros((num_times, num_points, num_points))
E_z_all = np.zeros((num_times, num_points, num_points))

# Loop over the time values
for t_index, t in enumerate(t_values):
    # Initialize electric field components
    E_z = np.zeros_like(Z)
    E_x = np.zeros_like(X)

    # Outside the fish (in air)
    E_z_outside = E0 * np.cos(k_air * Z - omega * t)
    E_x_outside = np.zeros_like(X)

    # Inside the fish
    E_amplitude_inside = E0 * np.exp(-alpha * distance_inside[inside_fish])

    # Initialize E_z_inside and E_x_inside
    E_z_inside = np.zeros_like(distance_inside[inside_fish])
    E_x_inside = np.zeros_like(distance_inside[inside_fish])

    # Only compute fields where theta_t is valid
    valid = np.abs(sin_theta_t) <= 1

    # Compute phase inside the fish
    E_phase_inside = k_fish * (X[inside_fish] * np.sin(theta_t) + Z[inside_fish] * np.cos(theta_t)) - omega * t

    # Assign electric field components where valid
    E_z_inside[valid] = E_amplitude_inside[valid] * np.cos(theta_t[valid]) * np.cos(E_phase_inside[valid])
    E_x_inside[valid] = E_amplitude_inside[valid] * np.sin(theta_t[valid]) * np.cos(E_phase_inside[valid])

    # Combine inside and outside fields
    E_z[~inside_fish] = E_z_outside[~inside_fish]
    E_x[~inside_fish] = E_x_outside[~inside_fish]
    E_z[inside_fish] = E_z_inside
    E_x[inside_fish] = E_x_inside

    # Store the computed fields for later use
    E_x_all[t_index] = E_x
    E_z_all[t_index] = E_z


```

```python
power_inside_all = np.zeros((num_times, num_points, num_points))
for t_index in range(num_times):
    Ex = E_x_all[t_index]
    Ez = E_z_all[t_index]
    power_inside = np.zeros_like(Ez)
    E_total_squared = Ex**2 + Ez**2
    P_r = 2 * np.pi * (c / lamb) * epsilon_0 * l_f * E_total_squared
    power_inside[inside_fish] = P_r[inside_fish]
    power_inside_all[t_index] = power_inside


# Calculate the total power absorbed by the fish at each time step
dx = x[1] - x[0]
dz = z[1] - z[0]
total_power_absorbed_time = np.zeros(num_times)
for t_index in range(num_times):
    total_power_absorbed_time[t_index] = np.sum(power_inside_all[t_index][inside_fish]) * dx * dz

# Calculate cumulative energy absorbed over time
energy_absorbed = np.cumsum(total_power_absorbed_time * (t_values[1] - t_values[0]))  # in Joules
```

```python
c_p = 4200  # J/(kgK)
m_fish = 4.5  # kg

# Initial temperature
T_initial = 20   # C

# Calculate temperature over time
Delta_T = energy_absorbed / (m_fish * c_p)
T_values = T_initial + Delta_T

# Plot temperature over time
plt.figure(figsize=(10, 6))
plt.plot(t_values, T_values)
plt.xlabel('Time (s)')
plt.ylabel('Temperature (C)')
plt.title('Temperature of the Fish Over Time')
plt.grid(True)
plt.show()
```

```python
dt = (t_values[-1] - t_values[0])
dT = (T_values[-1] - T_values[0])

new_dt = dt * 1e8
new_dT = dT * 1e8

new_t_vals = np.arange(0, 100, new_dt)
new_T_vals = np.zeros_like(new_t_vals)

new_T_vals[0] = T_values[0]
T = T_values[0]

for i in range(len(new_t_vals) - 1):
    T += new_dT
    new_T_vals[i+1] = T

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(new_t_vals, new_T_vals, label='Interpolated T values')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (C)')
plt.title('Temperature of the Fish Over Time')
plt.grid(True)
plt.show()
```
