# How to Kill a Fishingbird

## Introduksjon

Vi skal utforske hvordan en mikrobølgeovn fungerer i elektromagnetisk kontekst. Dette er spesifikt for å kunne se hvor raskt og hvordan en mikrobølgeovn kunne drept en fisk avhengig av størrelse og utforming på fisken
For å gjøre eksperimentet mer humant skal vi anta at fisken er sedert og dermed ikke ved bevissthet. Dette gjør det i tillegg lettere ettersom legemet vil ligge i ro under prosessen. 
Ettersom vi er i Norge er det ingen fisk mer passende til å representere oss enn laksen. 

## Teori

Gjennomsnittet av tall funnet på nettet med litt buffer her og der gir 22 minutter (1320 sek) overlevelsestid for en laks på land. Ettersom generelle tall er vanskelig å finne, sier vi at den gjennomsnittlige laksen overlever i den gjennomsnittlige tiden, og dette avtar/øker lineært avhengig av størrelsen på fisken, da bestemt  av vekt.

I de kommende likningene vil disse uttrykkene bli representert av tilhørende bokstaver:

$$
\text{Levetid} = t
$$ 
$$
\text{Størrelse\_på\_fisk} = m
$$
$$
\text{Konstant} = k, \;\;\;\;\;\text{$k$ er forholdet mellom størrelse og levetid}
$$

Med dette kan vi beskrive overlevelsestiden ved hjelp av likningen:

$$
t = k \cdot m
$$

Vi beregner da overlevelseskonstanten $k$ ved å snu på formelen, og får:

$$
k = \frac{\text{Levetid}}{\text{størrelse\_på\_fisk}} = \frac{t}{m} = \frac{1320 \, \text{sek}}{4.5 \, \text{kg}} = \frac{880}{3} \, \frac{\text{sek}}{\text{kg}}
$$

Vi antar dermed at den gjennomsnittlige laksen, omtrent 4.5 kg, overlever i 22 minutter, og at når størrelsen avtar, avtar også overlevelsestiden.

$$
\text{Levetid} = t = k \cdot m 
$$

Dermed vil en fisk på 30 kg være antatt å overleve i 8800 sek eller 146.66 minutter.

Dette er ved ganske ideelle leveforhold i forhold til en mikrobølgeovn. Derfor antar vi at en fisk i mikrobølgeovn vil dø $\sqrt{t}$ fortere enn forholdene brukt tidligere. Altså for en fisk på 4.5 kg:

$$
\text{ovn\_levetid} = \sqrt{t} = \sqrt{1320} \approx 36.3 \, \text{sek}
$$

Til sammenligning, fisk på 30 kg:

$$
\text{ovn\_levetid} = \sqrt{8800} \approx 93.8 \, \text{sek}
$$

Det var nok om fisken. Nå til hvordan en mikrobølgeovn fungerer.

Mikrobølgeovnen bruker oftest en frekvens på 2.45 GHz, og ovnen vi ser for oss har varmer fisken med en effekt på mellom 600 W til 1000 W, så dette er området vi ønsker å undersøke.

#### Modellering av fisken

Vi antar at laksen er en ellipsoide hvor lengden $2a = 35$ cm, bredden er $2b = 7$ cm og høyden er $2c = 5$ cm. Det gir oss verdiene vi trenger for følgende formel

$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2}  = 1
$$

#### Utregning for når fisken dør av mikrobølgeovnen    

Vi antar at bølgene varmer utenfra og inn, altså at den ikke varmer på innsiden før utsiden er varm. Vi antar også at en fisk kan sees som død dersom $\frac{1}{10}$ av fiskenes bredde er varmet. Etter litt søking her og der lander den gjennomsnittlige bredden på en norsk laks på omtrent 14 cm, som vi skal anta er nøyaktig heretter. Det vil si at når mikrobølgeovnen har varmet fisken $1.4 \, \text{cm}$ inn i kjøttet, vil vi anse fisken som død. Fisken starter på en temperatur på 10 grader Celsius, og vi anses dødelig temperatur som 40 grader Celsius. Dvs. fisken må varmes $30$ grader for å være dødelig.

En trenger litt under $4200 \, \text{J}$ ($4186 \, \text{J}$) for å varme 1 kg vann med 1 grad. Siden fisken lever i vann, vil massetettheten ikke være helt ulik vann, men den er litt tettere. For å gjøre livet vårt enklere, antar vi dermed at vi trenger $4200 \, \text{J}$ for å varme 1 kg fisk med en grad, eller $126{,}000 \, \text{J}$ for å varme 1 kg fisk til 40 grader, altså dødelig temperatur.

Nå gjenstår det kun å regne ut hvor mange kg $\frac{1}{10}$ av fiskens bredde tilsvarer.

##### Hvordan fungerer mikrobølgeovn
En mikrobølgeovn opererer ved hjelp av en magnetron som omdanner elektrisk energi til bølger. Det gjør den ved å få elektroner til å bevege seg i sirkulære baner som lager et indusert oscillerende magnetisk felt. Dette feltet vil generere mikrobølger som vi kan bruke til varme. 

$$
\lambda = \frac{c}{f}
$$


#### Elektrisk felt

Dielektriske egenskaper:
I lufta fra mikrobølgeovnen til den når fisken vil det være noe tap av styrke. Vi lar mikrobølgeovnen emitere et felt med styrke $E_0$ som går gjennom lufta som vi anser som vakuum og ser derfor for oss at vi har lite tap av elektrisk felt. Da har vi at $E(z) = E_0$ før bølgene har truffet fisken.

I fisken:
Vi har en dielektrisk konstant ($\epsilon_r$) inne i fisken som måler evnen til å ta opp elektrisk energi. \
Vi har også en tapfaktor kjent som $\tan{\delta}$ som representerer hvor effektivt materialet omdanner elektrisk energi til varme.

Siden vi antar at fisken består av mye vann så bruker vi dielektrisk konstant og tapfaktor til vann. For å forenkle bruker vi empiriske verdier ved romtemperatur ($20^\circ - 25^\circ$). Det er gitt ved $\epsilon_r = 78$ og $\tan\delta = \frac{Lf}{\epsilon_r}$, hvor $Lf$ er tapsfaktor (Loss factor) ofte gitt som $\epsilon''$ og er den imaginære delen av av den dielektriske konstanten. Dette har vi at er omtrent $Lf = 12$ fra empirisk data ved 2.45 GHz. Det gir $\tan\delta = \frac{12}{78} = 0.153846$ https://water.lsbu.ac.uk/water/microwave_water.html

Elektrisk felt:
Styrken på feltet vil da kun avhenge av hvor langt det har brutt inn i fisken som vi gir ved $r$. Dersm $E_0$ er styrken ved $r=0$ på overflaten får vi følgende:

$$
E(r) = E_0e^{-\alpha r}
$$
Hvor $\alpha$ er attenueringskonstanten gitt ved:

$$\alpha = \frac{\sqrt{2\epsilon_r}\pi f \tan{\delta}}{c} = \sqrt{2}\pi \frac{l_f}{\lambda\sqrt{\epsilon_r}}$$


Grunnen til at vi bruker $r$ nå og ikke $z$ er at feletet vil nå endre retning ettersom det endrer permitivitet og fisken er formet som en ellipsoide som gjør at det vil danne en innfalsvinkel. Dermed vil ikke feltet lenger virke kun i $z$-retning.



#### Oppvarming

Vi ønsker å finne $P$ som er energien som blir omdannet til varme. Den er gitt ved:
$$ P = 2 \pi f \epsilon_0 l_f E^2(r) = 2\pi \frac{c}{\lambda}\epsilon_0 l_f E_0^{2}e^{-\sqrt{2}\pi \frac{l_f}{\lambda\sqrt{\epsilon_r}}r}$$
 Som vi ser så er vi blitt tildelt et vakkert stykke som er behagelig for øynene, men om man ser litt nærmere etter ser vi at det kun er noen få ting vi  mangler. 
 



## Antagelser under modellering

Når mikrobølgeovnen er på vil platen fisken ligger på rotere. For å forenkle systemet mest mulig vil vi anta at mikrobølgeovnen kun avslutter etter hele rotasjoner av fisken, varmen da vil få en bedre form for symmetri over fisken. Vi antar også at mikrobølgeovnen er en prisme som emiterer konstante mikrobølger. 

Mikrobølgeovnen lar vi ha en kvadratisk bunn med sider $40$ cm og høyde $10$ cm.

```python
import numpy as np
from scipy import constants
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets
from IPython.display import display
```

```python
mu_0 = constants.mu_0
epsilon_0 = constants.epsilon_0
c = 3e8   
```

```python
# mikrobølgeovn
f = 2.45e9 # frekvens på mikrobølger 2.45 GHz
lamb = c / f # bølgelengde på mikrobølger
P = 800 # effekt
A = 0.4 * 0.4 # areal av toppen
T = 1 / f
omega = 2 * np.pi * f  # vinkelhastighet


E0 = np.sqrt((2*P) / (epsilon_0 * c* A))
E0
```

```python
#fisk
epsilon_r = 78
l_f = 12
tan_delta = l_f / epsilon_r
alpha = omega * np.sqrt(mu_0 * epsilon_0 * epsilon_r / 2) * tan_delta
delta = 1 / alpha
```

```python
r = np.linspace(0, 0.10, 500)  
E = 1 * np.exp(-alpha * r)

plt.figure(figsize=(10,6))
plt.plot(r*100, E)
plt.title('Relativ Elektrisk Feltstyrke inne i fisken')
plt.xlabel('Dybde (cm)')
plt.ylabel('Elektrisk Feltstyrke (V/m)')
plt.grid(True)
plt.show()
```

```python
# definere ellipsoide
a_ellipsoid = 0.35 / 2    
b_ellipsoid = 0.07 / 2
c_ellipsoid = 0.05 / 2
```

```python
num_points = 200

# polar koordinater - meshgrid
u = np.linspace(0, 2 * np.pi, num_points)
v = np.linspace(0, np.pi, num_points)
U, V = np.meshgrid(u, v)

X = a_ellipsoid * np.cos(U) * np.sin(V)
Y = b_ellipsoid * np.sin(U) * np.sin(V)
Z = c_ellipsoid * np.cos(V)

# plot ellipsoide
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='pink', alpha=0.6, edgecolor='pink')

# størrelse på mikrobølgeovn
ax.set_xlim([-0.2, 0.2]) 
ax.set_ylim([-0.2, 0.2])
ax.set_zlim([0, 0.1])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('Ellipsoid Plot')
plt.show()
```

```python
    # nye konstanter

mu_r = 1.0 # Fisk er ikke magnetisk

# Refraktiv indeks til bruk for Snells lov
n_air = 1.0
n_fish = np.sqrt(epsilon_r * mu_r)
k_air = omega / c  
k_fish = n_fish * omega / c  
```

```python

t0 = 0
t1 = 0.5 * T

num_points = 200
x = np.linspace(-0.20, 0.20, num_points)
z = np.linspace(0, 0.10, num_points)
X, Z = np.meshgrid(x, z)
Y = 0  
R_ellipsoid = (X / a_ellipsoid)**2 + (Y / b_ellipsoid)**2 + ((Z - c_ellipsoid) / c_ellipsoid)**2
inside_fish = R_ellipsoid <= 1

# Calculate the distance from the surface along the Z-axis
distance_inside = np.zeros_like(X)
Z_surface = c_ellipsoid + c_ellipsoid * np.sqrt(1 - (X[inside_fish] / a_ellipsoid)**2)
distance_inside[inside_fish] = Z_surface - Z[inside_fish]
distance_inside = np.maximum(distance_inside, 0)

# Initialize electric field components
E_z = np.zeros_like(Z)
E_x = np.zeros_like(X)

# Calculate the normal vector components at the surface
n_x = np.zeros_like(X)
n_z = np.zeros_like(Z)
n_x[inside_fish] = X[inside_fish] / a_ellipsoid**2
n_z[inside_fish] = (Z_surface - c_ellipsoid) / c_ellipsoid**2

# Calculate the angle of incidence and transmission using Snell's Law
# Calculate the angle of incidence using the normalized normal vectors
theta_i = np.arctan2(n_x[inside_fish], n_z[inside_fish])

# Calculate the sine of the transmission angle using Snell's Law
sin_theta_t = (n_air / n_fish) * np.sin(theta_i)

# Initialize theta_t with NaNs
theta_t = np.full_like(theta_i, np.nan)

# Identify where total internal reflection does not occur
valid = np.abs(sin_theta_t) <= 1

# Calculate theta_t where valid
theta_t[valid] = np.arcsin(sin_theta_t[valid])

# Loop over times t0 and t1
t_values = [t0, t1]

for t in t_values:
    # Outside the fish (in air)
    E_z_outside = E0 * np.cos(k_air * Z - omega * t)
    E_x_outside = np.zeros_like(X)

    # Inside the fish
    E_amplitude_inside = E0 * np.exp(-alpha * distance_inside[inside_fish])

    # Initialize E_z_inside and E_x_inside
    E_z_inside = np.zeros_like(distance_inside[inside_fish])
    E_x_inside = np.zeros_like(distance_inside[inside_fish])

    # Only compute fields where theta_t is valid
    valid = np.abs(sin_theta_t) <= 1

    # Compute phase inside the fish
    E_phase_inside = k_fish * (X[inside_fish] * np.sin(theta_t) + Z[inside_fish] * np.cos(theta_t)) - omega * t

    # Assign electric field components where valid
    E_z_inside[valid] = E_amplitude_inside[valid] * np.cos(theta_t[valid]) * np.cos(E_phase_inside[valid])
    E_x_inside[valid] = E_amplitude_inside[valid] * np.sin(theta_t[valid]) * np.cos(E_phase_inside[valid])


    E_z[~inside_fish] = E_z_outside[~inside_fish]
    E_x[~inside_fish] = E_x_outside[~inside_fish]
    E_z[inside_fish] = E_z_inside
    E_x[inside_fish] = E_x_inside

    # Compute the magnitude and normalize for quiver plot
    magnitude = np.sqrt(E_x**2 + E_z**2)
    Ex_norm = E_x / magnitude
    Ez_norm = -E_z / magnitude  # Adjust direction for plotting

    # Plotting the electric field strength at time t
    plt.figure(figsize=(10, 8))
    contour = plt.contourf(X, Z, E_z, levels=50, cmap="viridis")
    plt.colorbar(contour, label="Electric Field Strength (V/m)")
    plt.xlabel("X (m)")
    plt.ylabel("Z (m)")
    plt.title(f"Electric Field Distribution and Direction at t = {t:.2e} s")

    # Plot the ellipsoid boundary
    theta_boundary = np.linspace(0, 2 * np.pi, 300)
    x_ellipsoid = a_ellipsoid * np.cos(theta_boundary)
    z_ellipsoid = c_ellipsoid + c_ellipsoid * np.sin(theta_boundary)
    plt.plot(x_ellipsoid, z_ellipsoid, color='white', linewidth=2, label="Fish")

    # Plot electric field direction using quiver
    n_quiver = 20
    plt.quiver(X[::n_quiver, ::n_quiver], Z[::n_quiver, ::n_quiver],
               Ex_norm[::n_quiver, ::n_quiver], Ez_norm[::n_quiver, ::n_quiver],
               color='white', pivot='mid', scale=25, headwidth=3, headlength=4)

    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
```

```python
def P_total(Ex, Ez):
    epsilon_r_imag = epsilon_r * tan_delta
    E_total_squared = Ex**2 + Ez**2
    P_r = 0.5 * omega * epsilon_0 * epsilon_r_imag * E_total_squared
    return P_r
```

```python
power_outside = np.zeros_like(Z)

power_inside = np.zeros_like(X)
power_inside[inside_fish] = P_total(E_x_inside, E_z_inside)

plt.figure(figsize=(8, 6))
power_contour = plt.contourf(X, Z, power_inside, levels=50, cmap="inferno")
plt.colorbar(power_contour, label="Power Inside Fish (W)")
plt.xlabel("X (m)")
plt.ylabel("Z (m)")
plt.title("Power Distribution Inside the Fish")
plt.plot(x_ellipsoid, z_ellipsoid, color='white', linewidth=2, label="Fisk")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# LAGE MASSE FELT :D


# Generate 100 time values from 0 to T
t_values = np.linspace(0, T, 100)
num_times = len(t_values)

# Spatial grid parameters
num_points = 200
x = np.linspace(-0.20, 0.20, num_points)
z = np.linspace(0, 0.10, num_points)
X, Z = np.meshgrid(x, z)
Y = 0

# Define the ellipsoid representing the fish
R_ellipsoid = (X / a_ellipsoid)**2 + (Y / b_ellipsoid)**2 + ((Z - c_ellipsoid) / c_ellipsoid)**2
inside_fish = R_ellipsoid <= 1

# Calculate the distance from the surface along the Z-axis
distance_inside = np.zeros_like(X)
Z_surface = c_ellipsoid + c_ellipsoid * np.sqrt(1 - (X[inside_fish] / a_ellipsoid)**2)
distance_inside[inside_fish] = Z_surface - Z[inside_fish]
distance_inside = np.maximum(distance_inside, 0)

# Calculate the normal vector components at the surface
n_x = np.zeros_like(X)
n_z = np.zeros_like(Z)
n_x[inside_fish] = X[inside_fish] / a_ellipsoid**2
n_z[inside_fish] = (Z_surface - c_ellipsoid) / c_ellipsoid**2

# Calculate the angle of incidence and transmission using Snell's Law
theta_i = np.arctan2(n_x[inside_fish], n_z[inside_fish])
sin_theta_t = (n_air / n_fish) * np.sin(theta_i)

# Initialize theta_t with NaNs
theta_t = np.full_like(theta_i, np.nan)

# Identify where total internal reflection does not occur
valid = np.abs(sin_theta_t) <= 1

# Calculate theta_t where valid
theta_t[valid] = np.arcsin(sin_theta_t[valid])

# Pre-allocate arrays to store E_x and E_z at each time step
E_x_all = np.zeros((num_times, num_points, num_points))
E_z_all = np.zeros((num_times, num_points, num_points))

# Loop over the time values
for t_index, t in enumerate(t_values):
    # Initialize electric field components
    E_z = np.zeros_like(Z)
    E_x = np.zeros_like(X)

    # Outside the fish (in air)
    E_z_outside = E0 * np.cos(k_air * Z - omega * t)
    E_x_outside = np.zeros_like(X)

    # Inside the fish
    E_amplitude_inside = E0 * np.exp(-alpha * distance_inside[inside_fish])

    # Initialize E_z_inside and E_x_inside
    E_z_inside = np.zeros_like(distance_inside[inside_fish])
    E_x_inside = np.zeros_like(distance_inside[inside_fish])

    # Only compute fields where theta_t is valid
    valid = np.abs(sin_theta_t) <= 1

    # Compute phase inside the fish
    E_phase_inside = k_fish * (X[inside_fish] * np.sin(theta_t) + Z[inside_fish] * np.cos(theta_t)) - omega * t

    # Assign electric field components where valid
    E_z_inside[valid] = E_amplitude_inside[valid] * np.cos(theta_t[valid]) * np.cos(E_phase_inside[valid])
    E_x_inside[valid] = E_amplitude_inside[valid] * np.sin(theta_t[valid]) * np.cos(E_phase_inside[valid])

    # Combine inside and outside fields
    E_z[~inside_fish] = E_z_outside[~inside_fish]
    E_x[~inside_fish] = E_x_outside[~inside_fish]
    E_z[inside_fish] = E_z_inside
    E_x[inside_fish] = E_x_inside

    # Store the computed fields for later use
    E_x_all[t_index] = E_x
    E_z_all[t_index] = E_z


```

```python
power_inside_all = np.zeros((num_times, num_points, num_points))
for t_index in range(num_times):
    Ex = E_x_all[t_index]
    Ez = E_z_all[t_index]
    power_inside = np.zeros_like(Ez)
    E_total_squared = Ex**2 + Ez**2
    P_r = 2 * np.pi * (c / lamb) * epsilon_0 * l_f * E_total_squared
    power_inside[inside_fish] = P_r[inside_fish]
    power_inside_all[t_index] = power_inside


# Calculate the total power absorbed by the fish at each time step
dx = x[1] - x[0]
dz = z[1] - z[0]
total_power_absorbed_time = np.zeros(num_times)
for t_index in range(num_times):
    total_power_absorbed_time[t_index] = np.sum(power_inside_all[t_index][inside_fish]) * dx * dz

# Calculate cumulative energy absorbed over time
energy_absorbed = np.cumsum(total_power_absorbed_time * (t_values[1] - t_values[0]))  # in Joules
```

```python
c_p = 4200  # J/(kg·K)
m_fish = 4.5  # kg

# Initial temperature
T_initial = 20   # °C

# Calculate temperature over time
Delta_T = energy_absorbed / (m_fish * c_p)
T_values = T_initial + Delta_T

# Plot temperature over time
plt.figure(figsize=(10, 6))
plt.plot(t_values, T_values)
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Temperature of the Fish Over Time')
plt.grid(True)
plt.show()
```

```python
dt = (t_values[-1] - t_values[0])
dT = (T_values[-1] - T_values[0])

new_dt = dt * 1e8
new_dT = dT * 1e8

new_t_vals = np.arange(0, 100, new_dt)
new_T_vals = np.zeros_like(new_t_vals)

new_T_vals[0] = T_values[0]
T = T_values[0]

for i in range(len(new_t_vals) - 1):
    T += new_dT
    new_T_vals[i+1] = T

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(new_t_vals, new_T_vals, label='Interpolated T values')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Temperature of the Fish Over Time')
plt.grid(True)
plt.show()
```
